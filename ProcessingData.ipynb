```python
import pandas as pd

# Load the uploaded files
file_paths = {
    "movies_data": "D:/Escritorio/Second period 2024-2025/Data analysis/assigment 2 ntbk/movies_data_7100.csv",
    "movies_output": "D:/Escritorio/Second period 2024-2025/Data analysis/assigment 2 ntbk/movies_output_2025-01-15_21-13-47.csv",
    "movies_with_links": "D:/Escritorio/Second period 2024-2025/Data analysis/assigment 2 ntbk/movies_with_links.csv"
}

# Read the files into dataframes
dataframes = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Display the first few rows of each dataframe for inspection
dataframes_preview = {name: df.head() for name, df in dataframes.items()}
dataframes_shapes = {name: df.shape for name, df in dataframes.items()}

dataframes_preview, dataframes_shapes
```

```python
# Merge `movies_data` and `movies_with_links` on shared structure, adding unique columns
merged_data = pd.merge(dataframes['movies_data'], dataframes['movies_with_links'], on=['Title', 'ID', 'Rating', 'Release Year', 'Duration'], how='left')

# Integrate `movies_output` by aligning titles
final_data = pd.merge(merged_data, dataframes['movies_output'], on='Title', how='left')

# Inspect the resulting raw data
final_data_shape = final_data.shape
final_data_preview = final_data.head()

final_data_shape, final_data_preview
```

```python
final_data.to_csv("D:/Escritorio/Second period 2024-2025/Data analysis/assigment 2 ntbk/final_data.csv", index=False)
```

```python
# Function to clean movie titles (remove numbering and parentheses with year)
def clean_title(title):
    import re
    # Remove leading numbering and periods, and remove text in parentheses (e.g., "(2008)")
    return re.sub(r'^\d+\.\s*|\s\(\d{4}\)$', '', title).strip()

# Clean titles in all dataframes
dataframes['movies_data']['Cleaned_Title'] = dataframes['movies_data']['Title'].apply(clean_title)
dataframes['movies_with_links']['Cleaned_Title'] = dataframes['movies_with_links']['Title'].apply(clean_title)
dataframes['movies_output']['Cleaned_Title'] = dataframes['movies_output']['Title'].apply(clean_title)

# Merge using the cleaned titles
merged_data_corrected = pd.merge(
    dataframes['movies_data'], 
    dataframes['movies_with_links'], 
    on=['Cleaned_Title', 'ID', 'Rating', 'Release Year', 'Duration'], 
    how='left'
)

final_data_corrected = pd.merge(
    merged_data_corrected, 
    dataframes['movies_output'], 
    on='Cleaned_Title', 
    how='left'
)

# Inspect the corrected merged data
final_data_corrected_shape = final_data_corrected.shape
final_data_corrected_preview = final_data_corrected.head()

final_data_corrected_shape, final_data_corrected_preview
```

```python
final_data_corrected
```

```python
# Remove rows with any missing (NaN) values
final_data_cleaned = final_data_corrected.dropna()

# Inspect the cleaned data
final_data_cleaned_shape = final_data_cleaned.shape
final_data_cleaned_preview = final_data_cleaned.head()

final_data_cleaned_shape, final_data_cleaned_preview
```

```python
final_data_cleaned
```

```python
final_data_cleaned = final_data_cleaned.reset_index(drop=True)
final_data_cleaned['Index'] = final_data_cleaned.index + 1

final_data_cleaned = final_data_cleaned[['Index'] + [col for col in final_data_cleaned.columns if col != 'Index']]

final_data_cleaned_preview = final_data_cleaned.head()

final_data_cleaned_preview
```

```python
final_raw_data = final_data_cleaned.copy() 
output_raw_csv_path = "final_cleaned_movies_data_f.csv"
final_raw_data.to_csv(output_raw_csv_path, index=False, header=True)
```
